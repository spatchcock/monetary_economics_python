Here, we continue experimenting with the model structure established in the last few posts (here, here, here). Our purpose here will be to introduce some notion of long-term economic growth into our economy and to see how that effects the conclusions we've established so far.

In order to achieve this we will simply increase the level of government spending through time. We'll consider exactly what this increase might represent afterwards.

Okay, let's start with most of the standard stuff...

```python
# import required libraries
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

N = 100           # define number of time periods

# exogeneous variables

theta = 0.2       # tax rate

alpha_Y = 0.9     # propensity to consume out of income
alpha_H = 0.2     # propensity to consume out of saved wealth

# endogeneous variables

Y   = np.zeros(N) # income
T   = np.zeros(N) # tax revenue
C   = np.zeros(N) # consumption
H_h = np.zeros(N) # private savings
H_g = np.zeros(N) # government balance
```

Okay, now for the change. We want to increase government spending through time. So far, we've simply modelled government spending using constant values. Increasing spending simply involves representing $G$ not with a single, specified constant in our code, but as a time series of values which differ (in our case, increase) through time. In this model we'll go for the simplest possible change in government spending - we'll simply increase it linearly through time. This means we need is an equation of the form:

$$G = mt + c$$

where $G$ is government spending, $t$ is the time period, $c$ is the starting value and $m$ is the rate at which $G$ changes through time. In order to be consistent with previous models we'll start at a level of government spending of 20 pounds and therefore $c$ will equal 20. And, arbitrarily, we'll attempt to hit $G$ = 40 after the normal 100 time periods. This means we need to increase $G$ by $1/5$ of a pound in each time period. Therefore:

$$G = \frac {t}{5} + 20$$

This is, mathematically, what we need to express our increasing government spending. 

So, first, we create an array to hold the values, initially populated by zeros. Then we populate it according to the expression above:

```python
G = np.zeros(N) 

G = np.arange(0,N)/5.0 + 20
```

Note that the code <code>np.arange(0,N)</code> generates a list (an array) of numbers from 0 to 100 and therefore the whoe expression is evaluated as if each of these numbers is individually substituted into our linear equation above - it produces an array of answers which becomes our model time series for $G$.

Okay, let's run the model.

```python
for t in range(0, N):
    
    # calculate total income for this time step (equation 1)
    Y[t] = (G[t] + alpha_H*H_h[t-1])/(1 - alpha_Y*(1-theta))
    
    # calculate the tax paid on income for this time step (3)
    T[t] = theta * Y[t]
    
    # calculate the consumption spending for this time step (4)
    C[t] = alpha_Y*(1 - theta)*Y[t] + alpha_H*H_h[t-1]
    
    # calculate the new level of private savings for this time step (5)
    H_h[t] = H_h[t-1] + Y[t] - T[t] - C[t]
    
    # calculate the new level of government money balance (6)
    H_g[t] = H_g[t-1] + T[t]- G[t]
   
```

Notice that <code>G[t]</code>is now accessed as an array with time varying (<code>[t]</code>) elements.

```python
# initialise plot figure
fig = plt.figure(figsize=(12, 4))

# plot government spending (G) through time
gov_plot = fig.add_subplot(131, xlim=(0, N), ylim=(0, 200))         # set axis limits
gov_plot.plot(range(N), G, lw=3)                       # plot constant G versus time
gov_plot.grid()                                                     # add gridlines
plt.xlabel('time')                                                  # label x axis
plt.ylabel('government spending')                                   # label y axis

# plot consumption spending (C) through time
consumption_plot = fig.add_subplot(132, xlim=(0, N), ylim=(0, 200)) # set axis limits
consumption_plot.plot(range(N), C, lw=3)                            # plot C versus time
consumption_plot.grid()                                             # add gridlines
plt.xlabel('time')                                                  # label x axis
plt.ylabel('consumption')                                           # label y axis

# plot aggregate income (Y) through time
income_plot = fig.add_subplot(133, xlim=(0, N), ylim=(0, 200))      # set axis limits
income_plot.plot(range(N), Y, lw=3)                                 # plot Y versus time
income_plot.grid()                                                  # add gridlines
plt.xlabel('time')                                                  # label x axis
plt.ylabel('income')                                                # label y axis

plt.tight_layout() # space subplots neatly
```

Okay, so government spending indeed increases linearly from 20 pounds per time period at the start to 40 pounds per time period at the end. Consumption spending undergoes a fairly rapid increase during the first 10-20 time periods before settling into a phase of comparatively slower, but constant growth. Likewise, aggregate income initially grows rapidly and subsequently stabilises into constant, linear growth. 

The intial period of rapid increase in consumption spending and aggregate income looks similar to that seem in previous models as the economy establishes its preferred balance between spending and saving. The subsequent period of constantly increasing consumption and income is notably different from earlier models. What is interesting is that the lines do not represent a constant *value* but they do increase at a constant *rate*.

Let's look at the sectoral balances.

```python
# initialise plot figure
fig = plt.figure(figsize=(8, 4))

budget_plot = fig.add_subplot(121, xlim=(1, N), ylim=(-10, 10))                     # set axis limits
budget_plot.plot(range(N), T-G, lw=3)                                  # plot gov budget versus time
budget_plot.plot(range(N), Y-T-C, lw=3)                                             # plot private budget versus time
budget_plot.grid()                                                                  # add gridlines
plt.xlabel('time')                                                                  # label x axis
plt.ylabel('budget position')                                                       # label y axis

balance_plot = fig.add_subplot(122, xlim=(1, N), ylim=(np.min(H_g), np.max(H_h)))   # set axis limits
balance_plot.plot(range(N), H_g, lw=3)                                              # plot gov balance versus time
balance_plot.plot(range(N), H_h, lw=3)                                              # plot private balance versus time
balance_plot.grid()                                                                 # add gridlines
plt.xlabel('time')                                                                  # label x axis
plt.ylabel('money balance')                                                         # label y axis

plt.tight_layout() # space subplots neatly
```

As with earlier models, the private sector starts off in a position of surplus (green,left plot) which means that they are saving. In the first 10-20 time periods this surplus reduces. However, contrary to previous models, the private sector surplus never quite reduces to zero. Similarly, the government's budget deficit - which is the mirror-image of the private sector surplus - decreases but never reaches zero, the government doesn't quite reach a balanced budget. Instead, for most of the model run period, the private and government sectors have equal and opposite, constant levels of surplus and deficit respectively. This translates into constantly increasing private sector saved wealth and government debt (right plot).

## Steady-state in a growing economy
So we have an economy that is continually growing. As government spending increases, this drives an increase in induced consumption spending and continous growth in aggregate incomes. This does not look like the steady-state conditions that we've seen in previous models where most or all of our macroeconomic variables (e.g. $C$, $T$, $Y$, $H_h$, $H_g$) reached constant levels. Instead, each of these increases with each time period, driven by the increase in government spending.

However, if we look at the size of saved wealth relative to aggregate income - the wealth/income ratio - we can understand what's going on a bit more clearly.

```python
# initialise plot figure
fig = plt.figure(figsize=(8, 4))

wealth_ratio_plot = fig.add_subplot(121, xlim=(0, N), ylim=(0,0.6))      # set axis limits
wealth_ratio_plot.plot(range(0,N), H_h/Y, lw=3)                          # plot wealth:income versus time
wealth_ratio_plot.grid()                                                 # add gridlines
plt.xlabel('time')                                                       # label x axis
plt.ylabel('saved wealth:income ratio')                                  # label y axis                                                       # label y axis

plt.tight_layout() # space subplots neatly
```

Here, we see that the level of saved wealth relative to income initially increases but eventually levels of at a value close to 0.4 - that is saved wealth is approximately 40% of aggregate income. *This* aspect of our model is similar to previous ones. So although we don't have a steady-state model in the sense of enirely constant values, we are seeming some form of stability in the wealth/income ratio.

Recall that in previous posts we considered the steady-state wealth/income ratio to represent the savings target of our population. Well in this case, the model does converge to a stable savings target (relative to income), it is just that income is continually increasing and therefore saved wealth increases continually in correspondence with this target. We can, in some sense, consider this to be a stable economy, therefore. We have growth in spending and incomes, but a stable wealth to income ratio. Saved wealth is stable relative to the *size of the economy*. 

Godley and Lavoie (2007) differentiate between steady-state behaviour of this kind and the kind described previously. Where all macroeconomic variables simply exhibit constant values, they use the term *stationary state*. So under this nomenclature what we have seen up till now have all been *stationary-state* models. Models which display some dynamism in their variables (such as continual growth) but exhibit stability in other characteristics (e.g. wealth to income ratio) are described by Godley and Lavoie (2007) with the term we have used up till now, *steady-state*. So the present model is our first actual "steady-state" model, in this sense.

In growing economies we obviously find that particular economic variables increase through time and this makes comparisons through time complicated. It is often convenient to consider such variables in relation to some other factor such as the size of the economy at the time. This approach adjusts for the underlying context of growth and enables us to understand whether anything is fundamentally changing in the economy aside from everything just getting bigger. So we can consider this to be a steady-state economy in that when we strip away the aspect of growth we have stable variables.

## What does this growth represent
It is useful to consider what the growth in our model might represent. Remember, all we did was to increase government spending through time. So what exactly could this be argued to mean? Well, our government is a sovereign currency issuing government. It spends by issuing IOUs to the private and these IOUs circulate as money. It is no problem for our government to be increasing the amount it spends from a *financing* perspective, but we do need to consider what the effects on the economy might be.

Economic growth is usually described in terms of *real* or *nominal* growth. Real economic growth represents an increase in the goods and services that have been produced. This may result from a simple increase in population - more people around to produce things - and/or an increase in *productivity* - the ability to produce more with the same resources.
