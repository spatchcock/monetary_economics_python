The reason this economy stabilizes is down to the different ways in which government spending and taxation are defined: spending as an *absolute* quantity per time period (20 pounds) and tax *relative* to income ($20%$). Listen to any government budget announcement and this structure of spending and tax policy will sound familiar. Governments usually choose how many (billions of) pounds to spend on a variety of services but do *not* choose how much tax to collect (they may wish they could). Instead, they choose what *percentages* of different economic activities to tax. The actual amount collected depends therefore on the level of those economic activities. This difference between how spending and taxation are are defined is highly significant for the economy. It means that spending and tax impart a *feedback* on the economy which, in our model, goes something like this: 

1. a spending deficit causes the economy to grow
2. the growing economy causes tax revenue to increase
3. the increasing tax revenue causes the spending deficit to reduce
4. the decreasing deficit causes the economy to grow more slowly

This process repeats until the budget is balanced and the economy stable. This occurs specifically *because* spending is stipulated as an *absolute* amount whereas taxation is stipulated as a *relative* amount. It means that taxes grow as incomes grow and eventually reach the fixed level at which spending is set. At this point everything in the economy is stable, unchanging. This economy is inherently stable because of the dynamics of government spending and taxation. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

N = 100


G = 20

theta = 0.2
alpha = np.zeros(N)
alpha[0:50] = 0.98
alpha[50:] = 0.97

C   = np.zeros(N) # consumption
Y   = np.zeros(N) # income
Y_d = np.zeros(N) # disposable income
T   = np.zeros(N) # tax revenue
H_h = np.zeros(N) # private savings
H_g = np.zeros(N) # government debt

for t in range(0, N):
    
    # calculate consumer spending
    C[t]   = alpha[t]*Y_d[t-1] 
    
    # calculate total income (consumer spending plus constant government spending)
    Y[t]   = G + C[t] 
    
    # calculate the tax take
    T[t] = theta * Y[t]
    
    # calculate disposable income
    Y_d[t] = Y[t] - T[t]
    
    # calculate the change in private savings
    H_h[t] = H_h[t-1] + (1-alpha[t])*Y_d[t-1] 
    
    # calculate the change in government debt
    H_g[t] = H_g[t-1] + T[t]- G
   

```

```python
fig = plt.figure(figsize=(12, 8))

consumption_plot = fig.add_subplot(241, xlim=(0, N), ylim=(0, np.max(Y)))
consumption_plot.plot(range(N), C, lw=3)
consumption_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('consumption')

gov_plot = fig.add_subplot(242, xlim=(0, N), ylim=(0, np.max(Y)))
gov_plot.plot(range(N), np.repeat(G,N), lw=3)
gov_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('government spending')

income_plot = fig.add_subplot(243, xlim=(0, N), ylim=(0, np.max(Y)))
income_plot.plot(range(N), Y, lw=3)
income_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('income')

savings_plot = fig.add_subplot(244, xlim=(0, N), ylim=(0, np.max(H_h)))
savings_plot.plot(range(N), H_h, lw=3)
savings_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('private savings')

gov_plot = fig.add_subplot(245, xlim=(0, N), ylim=(0, np.max(G)*1.5))
gov_plot.plot(range(N), np.repeat(G,N), lw=3)
gov_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('government spending')

tax_plot = fig.add_subplot(246, xlim=(0, N), ylim=(0, np.max(G)*1.5))
tax_plot.plot(range(N), T, lw=3)
tax_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('tax revenue')

deficit_plot = fig.add_subplot(247, xlim=(0, N), ylim=(np.min(T-np.repeat(G,N))*1.5,0))
deficit_plot.plot(range(N), T-np.repeat(G,N), lw=3)
deficit_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('government budget')

debt_plot = fig.add_subplot(248, xlim=(0, N), ylim=(np.min(H_g)*1.5,0))
debt_plot.plot(range(N), H_g, lw=3)
debt_plot.grid()
# label axes
plt.xlabel('time')
plt.ylabel('government debt')

# space subplots neatly
plt.tight_layout()
```
