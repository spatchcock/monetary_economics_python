This post will build on the models developed so far by attempting to represent the concept of changing prices - that is price inflation (or deflation, which is simply negative inflation). Inflation has a number of consequences and we'll attempt to construct an accounting framework that adequately deals with some of them, albeit in a quite simplified manner.

What is inflation?
------------------
Firstly, lets clarify what we mean by "price inflation". The prices of specific goods, services or financial assets (such as shares or government bonds) may vary over time. Individually these might be quite unrelated and simply reflect the nature of the respective markets at given points in time. What we are concerned with here is a more general change in overall prices. This is usually measured using a reference "basket" of goods and services, taken as being representative of prices in general.

With the concept of changing prices, it becomes apparent that comparing economic quantities between different points in time may not be completely informative. In one sense, 100 pounds today is worth precisely 100 pounds in one year's time, irrespective of whether prices of goods have changed. This because the pound is the unit that monetary quantities are measured in (in the UK at least, the pound is the "unit of account"). It follows that, by definition, 100 pounds is always "worth" exactly 100 pounds. 

In another sense, however, the *value* of 100 pounds in terms of the goods and services that it can purchase may have changed. For example, if prices have doubled, 100 pounds today would only buy half of what it could buyone year ago. So if we consider the value of our money to be more properly represented by its *purchasing power*, we might be better off thinking of our 100 pounds as being worth just 50 pounds (relative to one year ago). 

Using the face value of economic quantities is described as the *nominal value* whereas adjusting these quantities to take into account price inflation is described as the *real value*. Macroeconomic quantities such as government spending, aggregate income, saved wealth, government debt, etc., can equally be described in terms of either nominal or real values. 

The price level
---------------
There are a few ways to deal with changing prices. We can think in terms of a relative *price level* or we can think in terms of a *rate of change* (i.e. the inflation rate). 

The price level is usually simply defined in terms of how much general prices have changed relative to an earlier time period. In this way we do not have to deal with absolute prices - which vary widely across individual goods, services and assets - but simply their relative change. More formally, the price level is not denominated in pounds (or what ever the unit of account is) but is a dimensionless quantity (that is, just a number) because it represents a quantity in pounds (this year) divided by another quantity in pounds (the reference year).

The relationship between the "price level" and the rate of inflation can be described as follows:  

$$\pi = \frac{\Delta p}{p_{t-1}}$$

where $\pi$ is the inflation rate, $p_t$ is the price level during a given time period $t$ and $\Delta p$ is the change in price level. A less succinct version would be:

$$\pi = \frac{p_{t} - p_{t-1}}{p_{t-1}} = \frac{p_{t}}{p_{t-1}} - 1$$

This expression enables us to calculate an inflation rate froma pair of price level values. A useful reformulation is:

$$p_t = (\pi+ 1) p_{t-1}$$

which allows us to calculate a price level from a measure of inflation.

Accounting for inflation
------------------------

In the last model the balance sheets of each sector looked something like this:

||household|firms|government|$\Sigma$|
|------|------|------|------|------|
|Money held|$+H_h$||$-H_g$|0|

This *balance sheet matrix* makes it explicit that the only financial asset in this economy is what is being considered straightforwardly as 'money'. This money is only held by one economic sector - households - and comprises what we refer to as 'saved wealth'. Further, this money is issued by, and is a liability of, the government. What we need to figure out is how inflation affects these stocks.

The first thing to note is that these quantities are *nominal* quantities. In general, flows and stocks of money are recorded as nominal values, but there will be some situations in which we need to deal in real values. We will continue to denote nominal values using capitalised variables and use lower case for their real counterparts, e.g. $Y$ for nominal aggregate income and $y$ for real aggregate income.

So we know that the nominal quantity of money held by the household sector at time $t$ is denoted $H_{h,t}$. We can now state that the real value of this stock is termed $h_{h,t}$ and that this is determined by "deflating" (i.e. dividing) the nominal quantity by the price level $p_t$, i.e.

$$h_{h,t} = \frac{H_{h,t}}{p_t}$$

The real value of saved wealth in the *previous* time period is defined similarly, but for a change in the subscripts:

$$h_{h,t-1} = \frac{H_{h,t-1}}{p_{t-1}}$$

In other words, the *current* real value of saved wealth is equal to the *current* nominal value divided by current prices. The previous real value of saved wealth was the *previous* nominal value divided by the *previous* price level. In general, the real value of saved wealth at any given point in time is equal to the nominal value of saved wealth at that time divided by the price level at that time.  

Now, suppose we have a given nominal value of existing wealth, e.g $H_{h,t-1}$, and then the price level changes. The nominal size of our wealth has not changed, but the value of that wealth, in terms of its purchasing power, has. We can describe the change in this real value by:

$$\Delta h_h = \frac{H_{h,t-1}}{p_{t}} - \frac{H_{h,t-1}}{p_{t-1}} $$

That is, the difference between the previous nominal value of wealth deflated by *previous* prices and the *same* nominal value of wealth deflated by *current* prices. With some maniputations we can arrive at,

$$\Delta h_h = \Big ( \frac{1}{p_{t-1}}-\frac{1}{p_{t}} \Big) H_{h,t-1}$$

$$\Delta h_h = \Big ( \frac{1 + \pi}{p_t}-\frac{1}{p_{t}} \Big) H_{h,t-1}$$

$$\Delta h_h = \Big ( \frac{1 + \pi - 1}{p_t} \Big) H_{h,t-1}$$

$$\Delta h_h =  -\pi \frac{H_{h,t-1}}{p_t}$$

This states that the change in the real value of household wealth is equal to the *previous nominal* value of wealth deflated by *current* prices and multiplied by the inflation rate. Note the leading minus sign, meaning that for a positive inflation rate, the change in real wealth is negative, i.e. a loss. 

Developing a model with inflation
---------------------------------

Recall that the previous model was based upon the following transactions matrix:

||household|firms|government|$\Sigma$|
|------|------|------|------|------|
|**flows**|||||
|Government spending||$+G$|$-G$|$0$|
|Payment of taxes|$-T$||$+T$|$0$|
|Income (wages)|$+Y$|$-Y$||$0$|
|Consumption spending|$-C$|$+C$||$0$|
||||||
|**stocks**|||||
|Change in money|-$\Delta H_h$||+$\Delta H_g$|0|
||||||
|**$\Sigma$**|$0$|$0$|$0$|$0$|

And the equations used to drive the model were specified as:

|||
|--|--|
|1. | $Y_t = C_t + G$|
|2. | $G = G$ (exogenous)|
|3. | $T_t = \theta T_t$|
|4. | $C_t = \alpha_1 (1 - \theta) Y_t + \alpha_2 H_{h,t-1}$|
|5. | $H_{h,t} = H_{h,t-1} + Y_t - C_t - T_t $|
|6. | $H_{g,t} = H_{g,t-1} + T_t - G$|

In this model spending was done by government and the private sector, with private sector spending determined by both income and saved wealth (from the previous time period). We need to make a few changes in order to accommodate real and nominal prices and the implications of price changes.

We will model changes in price as an exogenous process. That is, we will simply specify the levels of inflation at our own whim, and see how that affects our economy. In future models we might aspire to *endogenise* inflation, making it part of the dynamic processes that emerge within the economic system. For now, though, we will apply a change in prices at the start of each time period. 

So we'll start by introducing a couple of new equations. First, we'll assume that inflation is constant, e.g. 2%. So we have, simply,

$$\pi = \pi$$

i.e. wahtever value we asign. And if we're imposing this inflation at the start of each time period we need to adjust the price level, $p$. We'll set the initial price level to $1$ meaning that our first time period is the *reference* time period for the price level. And we'll change the price level at each time step according to:

$$p_{t} = p_{t-1} (1 + \pi)$$

So in each time period we have a new price level. The important implication of this is that the *real* value of any stocks of money held over from the previous time period need to change. We know how to describe these changes in real values, and so we'll step through the model logic and see how to incoporate this.

Now the big change in assumption we're going to make here is that folk decide on their spending activities based on real, rather than nominal, quantities. Previously we were implicitly assuming it was based on nominal quantities because nominal quantities were all we had in our model. It makes more sense to consider that folk are most likely to want to consume a certain amount of real goods and services, and therefore base their spending and saving decisions on that rather than have spending and saving targets defined in abstract numbers irrespective of prices. So we need to replace equation 4 - the "consumption function" - with something formulated in real quantities.

To do this we need to step back a little. Equation 4 is presented in a form that makes the system of 6 equations more succinct. A more general form for equation 4 for is:

$$C_t = \alpha_1 (Y_t - T_t) + \alpha_2 H_{h,t-1}$$

Notice that equation 4 is obtained by substituting equation 3 into the above expression. Now the term $Y_t - T_t$ can be thought of as *disposable income* - that is, the income left over after the payment of taxes. So we can describe an even more general form for the consumption function,

$$C_t = \alpha_1 YD_t + \alpha_2 H_{h,t-1}$$

where $YD_t = Y_t - T_t$. This equation says that folk spend a certain proportion of their *disposable income* and a certain proportion of their saved wealth in each given time period. All of this was implicit before, just not expressed explicitly.

Now, if nominal disposable income is the difference between nominal income and nominal taxes ($Y - T$) then we can understand the real value of this disposable income (the actual purchasing power of goods and services) by deflating this quantity by the current price level, i.e.

$$yd_t = \frac{YD_t}{p_t}$$

What we will state here, however, is that the above expression is incomplete. What we are missing is the change in real saved wealth that has occurred because of inflation. This might seem like the wrong place to consider this, but stay with it! 

Although saved wealth in this model is considered to be a *stock* variable, a change in saved wealth is equivalent to a *flow*. To understand this, consider that it is meaningless to consider a change in a stock quantity independent of the time over which the change occurred: was it in a single day, over a quarter, a year, what? A change in a stock quantity needs to be expressed with reference to particular points in time, and therefore this becomes a case of a monetary quantity divided by a time quantity, which is equivalent to a flow quantity. 

Another way to view this is through the lense of stock-flow consistency. In this framework, stocks and flows have to be rigorously related, stocks must arise only from flows, and flows must accumulate into stocks. A change in a stock quantity implies that a flow quantity is required somewhere in the framework. What we are doing here is treating the inflation loss as a flow between sectors. (This approach can be widened to refer to capital gains in general. When a financial asset appreciates in value, this is equivalent to a flow of income to the owner).

So while nominal disposable income is:

$$YD_t = Y_t - T_t$$

real disposable income requires an extra term:

$$yd_t = \frac{YD_t}{p_t} -\pi \frac{H_{h,t-1}}{p_t}$$

The extra term being the inflation loss that we derived above.

So the noteworthy implication of all of this logic is that real disposable income **does not** simply equal nominal disposable income deflated by current prices. More specifically, disposable real income (in this model) is lowered by the presence of inflation - folk have lower purchasing power once taxes, but also depreciation, are taken in to consideration.

Now we can finally define the consumption decision based on *real* income,

$$c_t = \alpha_1 yd_t + \alpha_2 \frac{H_{h,t-1}}{p_t}$$

With $yd_t$ defined as above. Notice how the second term - the spending out of saved wealth - now refers the *current* value of existing saved wealth: that is the *previous* nominal value of saved wealth deflated by *current* prices. 

Next, we'll specify a few more relations. First we'll say that the government also sets its spending level in real terms. So we have,

$g = g$ (exogenous)

But that this is simply related to the nominal government spend via:

$$G_t = g \cdot p_t$$

Other nominal quantities are also expressed similarly:

$$Y_t = y_t \cdot p_t$$

$$C_t = c_t \cdot p_t$$

$$T_t = t_t \cdot p_t$$

Finally we need to describe the change in monetary stocks for each sector. Nominally we have,

$$H_{h,t} = H_{h,t-1} + Y_t - C_t - T_t $$

$$H_{g,t} = H_{g,t-1} + T_t - G_t$$

And the real value at the end of a time period is simply these values deflated by the current price level,

$$h_{h,t} = \frac{H_{h,t}}{p_t}$$

$$h_{g,t} = \frac{H_{g,t}}{p_t}$$

So, to recap, we have:

|||
|--|--|
|1. | $\pi = \pi$ (exogenous)|
|2. | $p_t = p_{t-1} (1 + \pi)$|
|3. | $y_t = c_t + g$|
|4. | $g = g$ (exogenous)|
|5. | $c_t = \alpha_1 yd_t + \alpha_2 \frac{H_{h,t-1}}{p_t}$|
|6. | $yd_t = \frac{Y_t - T_t}{p_t} -\pi \frac{H_{h,t-1}}{p_t}$|
|7. | $T_t = \theta Y_t$|
|8. | $G_t = g \cdot p_t$|
|9. | $Y_t = y_t \cdot p_t$|
|10. | $C_t = c_t \cdot p_t$|
|11. | $T_t = t_t \cdot p_t$|
|12. | $H_{h,t} = H_{h,t-1} + Y_t - C_t - T_t $|
|13. | $H_{g,t} = H_{g,t-1} + T_t - G$|
|14. | $h_{h,t} = \frac{H_{h,t}}{p_t}$|
|15. | $h_{g,t} = \frac{H_{g,t}}{p_t}$|

These equations are sufficient to specify our entire economy. As with previous models though, we don't have a clear way forward because equation 3, which defines real income (or equivalently, real value of output) contains a reference to equation 9 (via equations 5 and 6) which itself refers to real income. These definitions are therefore circular and require a solution. We want to arrive at an alternative formulation of equation 3 which refers only to explicitly known quantities.

To start with, we'll substitute equation 7 into equation 6,

$yd_t = \frac{Y_t - \theta Y_t}{p_t} -\pi \frac{H_{h,t-1}}{p_t}$

and thus,

$yd_t = \frac{Y_t(1 - \theta)}{p_t} -\pi \frac{H_{h,t-1}}{p_t}$

Substituting equation 9 and cancelling the $p_t$ terms,

$yd_t = (1 - \theta) y_t -\pi \frac{H_{h,t-1}}{p_t}$

We can now substitute this expression for real disposable income into our consumption function, equation 5,

$$c = \alpha_1 \bigg [ (1 - \theta) y_t - \pi\frac{ H_{h,t-1}}{p_t} \bigg ] + \alpha_2 \frac{H_{h,t-1}}{p_t}$$
$$c = \alpha_1(1 - \theta)y_t - \alpha_1 \pi \frac{ H_{h,t-1}}{p_t} + \alpha_2 \frac{H_{h,t-1}}{p_{t}}$$
$$c = \alpha_1(1 - \theta)y_t + (\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t}$$

This now represents our effective consumption function, which defines consumption behaviour in real terms, as we wanted. It differs from previous versions in that the proportion of spending out of saved wealth is now slightly reduced by the amount $\alpha_1 \pi$.

Next we can substitute this expression into our equation for aggregate real income (equation 3):

$$y_t = \alpha_1(1 - \theta)y_t + (\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t} + g$$

$$y_t - \alpha_1(1 - \theta)y_t = (\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t} + g$$

$$y_t [1 - \alpha_1(1 - \theta)] = (\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t} + g$$

$$y_t = \frac{g +(\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t}}{1 - \alpha_1(1 - \theta)}$$


We can now list our final equations as:

|||
|--|--|
|1. | $\pi = \pi$| 
|2. | $p_t = p_{t-1} (1 + \pi)$|
|3. | $y_t = \frac{g +(\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t}}{1 - \alpha_1(1 - \theta)}$|
|4. | $g = g$ |
|5. | $c = \alpha_1(1 - \theta)y_t + (\alpha_2 - \alpha_1 \pi) \frac{H_{h,t-1}}{p_t}$|
|6. | $yd_t = \frac{Y_t - T_t}{p_t} -\pi \frac{H_{h,t-1}}{p_t}$|
|7. | $T_t = \theta Y_t$|
|8. | $G_t = g \cdot p_t$|
|9. | $Y_t = y_t \cdot p_t$|
|10. | $C_t = c_t \cdot p_t$|
|11. | $T_t = t_t \cdot p_t$|
|12. | $H_{h,t} = H_{h,t-1} + Y_t - C_t - T_t $|
|13. | $H_{g,t} = H_{g,t-1} + T_t - G_t$|
|14. | $h_{h,t} = \frac{H_{h,t}}{p_t}$|
|15. | $h_{g,t} = \frac{H_{g,t}}{p_t}$|


A model with inflation
----------------------

Okay let's run the model in the usual way, iterating through each time period and solving the respective equations. We'll start by setting the same default parameter values as previously:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 

N = 100

g     = 20  # real government spending
theta = 0.2 # tax rate

alpha_1 = 0.9 # propensity to spend out of income
alpha_2 = 0.2 # propensity to spend out of saved wealth
```

Next, we'll set the constant rate of inflation. We'll use a value of 2%.

```python
pi = 0.02
```

Next we'll initialise some containers for the various variables that we'll solve for. Note we now need variables for all of the real, as well as nominal, variables. And note we need a container for the price level variable, $p$.

```python
p   = np.zeros(N) # price level

y   = np.zeros(N) # real income
t   = np.zeros(N) # real tax revenue
c   = np.zeros(N) # real consumption
h_h = np.zeros(N) # real private savings
h_g = np.zeros(N) # real government balance

G   = np.zeros(N) # nominal government spending
Y   = np.zeros(N) # nominal income
T   = np.zeros(N) # nominal tax revenue
C   = np.zeros(N) # nominal consumption
H_h = np.zeros(N) # nominal private savings
H_g = np.zeros(N) # nominal government balance
```

The final thing we need to specify is the initial condition for the price level. We want this to be set to $1$ so that our first time period represents the reference year for adjusting prices.

```python
p[0] = 1
```

Now we can run our model. Note that we've switched the counter variable for the time period from <code>t</code> to <code>k</code>. This frees up the <code>t</code> variable to represent real tax revenues in line with our conventions for upper and lower case variable names.

```python
for k in range(1, N):

    p[k] = (1 + pi) * p[k-1]

    y[k] = (g + (alpha_2 + (alpha_2 - alpha_1)*pi)*H_h[k-1]/p[k])/(1 - alpha_1*(1-theta))
    c[k] = alpha_1*(1 - theta)*y[k] + (alpha_2 + (alpha_2 - alpha_1)*pi)*H_h[k-1]/p[k]
    
    G[k] = g*p[k]
    Y[k] = y[k]*p[k]
    T[k] = theta* Y[k]
    C[k] = c[k]*p[k]
    t[k] = T[k]/p[k]
    
    H_h[k] = H_h[k-1] + Y[k] - T[k] - C[k]
    H_g[k] = H_g[k-1] - G[k] + T[k]
    
    h_h[k] = H_h[k]/p[k]
    h_g[k] = H_g[k]/p[k]
    
```

Now let's plot aggregate income for starters.

```python
# initialise plot figure
fig = plt.figure(figsize=(8, 4))

income_plot = fig.add_subplot(121, xlim=(1, N), ylim=(0, np.max(Y)))                   
income_plot.plot(range(N), Y, lw=3)                                  
income_plot.plot(range(N), y, lw=3)                                          
income_plot.grid()   
income_plot.legend(['nominal', 'real'])

plt.xlabel('time')                                                                 
plt.ylabel('real versus nominal income')  

price_plot = fig.add_subplot(122, xlim=(1, N), ylim=(0, np.max(p)))                   
price_plot.plot(range(N), p, lw=3)                                  
price_plot.grid()   

plt.xlabel('time')                                                                 
plt.ylabel('price level') 

plt.tight_layout() # space subplots neatly
```

Okay, so we have two lines here, one for nominal aggregate income and one for real aggregate income. The real aggregate income looks similar to the pattern seen in previous models: an initial increase followed a stable, steady-state value of close to 100 pounds. Nominal incomes increase exponentially, reaching ~700 pounds after 100 time periods.

The trajectory of real incomes mimics the earlier models because we used the same behavioural parameter values as before (e.g. spending propensities, government spending and tax levels) but crucially, in this context, applied them in *real terms*, i.e. government and consumption spending behaviours were specified in real terms. It follows that the private sector wealth target was defined - in real terms - at a similar level as before, and therefore the steady-state condition was maintained with *real* income at a similar absolute level as before. So from this perspective we have a pretty similar economic scenario as before. 

The nominal values, are different, but really just reflect the price changes that we imposed. This can be gleaned from comparing the growth in nominal aggregate income with the change in the price level (left and right charts, blue lines). We can conclude from all this that, despite the impression given by the increasing nominal income curve, there isn't any real economic growth in this economy. The growth in nominal income simply represents the effects of inflation.

This gets us most of the way to understanding what is happening in this economy, but there are some important nuances. For example, real aggregate incomes are not exactly the same as before. Let's inspect our last real income value:

```python
y[-1]
```

96 pounds rather than 100 pounds. So aggregate real income has been reduced a little because of the presence of inflation. This arises from the reduction in real disposable income that inflation brings and which feeds into the decision to to engage in consumption spending. Consumption spending is reduced and therefore aggregate income is brought down comparatively.

Sectoral balances
-----------------

Before we look at the sectoral balances we need to figure out how to determine the real budget positions of each sector. We calculated the nominal wealth balances of each sector as follows:

$$H_{h,t} = H_{h,t-1} + Y_t - C_t - T_t $$

for the private sector, and,

$$H_{g,t} = H_{g,t-1} + T_t - G_t$$

for the government. From these equations, it is trivial to derive the *change* in balances which occurs during each time period. For example, the change in government debt is

$$\Delta H_{g} = H_{g,t} - H_{g,t-1} = T_t - G_t$$

This is what we understand to be the government's deficit (or, in principle, surplus) *in nominal terms*. Calculating the real terms deficit is slightly less trivial. Recall that we calculated the real terms value of the government debt in each time period as,

$$h_{g,t} = \frac{H_{g,t}}{p_t}$$

i.e. just the nominal value deflated by the price level. This can't be quite so easily rearranged in to an expression for the real deficit. Instead we need to make some changes. Firstly, we substitute the expression for the nominal government debt for the current period (i.e. described above):

$$h_{g,t} = \frac{H_{g,t-1} + T_t - G_t}{p_t}$$

Separate the terms on the right-hand side,

$$h_{g,t} = \frac{H_{g,t-1}}{p_t}+\frac{T_t}{p_t}-\frac{G_t}{p_t}$$

and simplify:

$$h_{g,t} = \frac{H_{g,t-1}}{p_t}+ t_t-g_t$$

Now, the first term represents the previous nominal debt deflated by current prices. We can write this in an alternative way, as follows:

$$h_{g,t} = H_{g,t-1} \frac{1}{p_t}+t_t-g_t$$

$$h_{g,t} = H_{g,t-1} \frac{1 + \pi - \pi}{p_t}+t_t-g_t$$

$$h_{g,t} = H_{g,t-1} \Bigg(\frac{1 + \pi}{p_t} - \frac{\pi}{p_t}\Bigg)+t_t-g_t$$

Using the definition $p_t = p_{t-1}(1 + \pi)$, we can write,

$$h_{g,t} = H_{g,t-1} \Bigg(\frac{1}{p_{t-1}} - \pi \frac{1}{p_t}\Bigg)+t_t-g_t$$

and then multiply through the $H_{g, t-1}$ term,

$$h_{g,t} = \frac{H_{g,t-1}}{p_{t-1}} - \pi \frac{H_{g,t-1}}{p_t}+t_t-g_t$$

The first term on the right hand side is now simply the *real* government debt from the last time period, and the second term is the inflation loss term which we derived above. So,

$$h_{g,t} = h_{g,t-1} + t_t - g_t - \pi \frac{H_{g,t-1}}{p_t}$$

Now we are in a position write explicitly what the change in real government debt is:

$$\Delta h_{g} = h_{g,t} - h_{g,t-1} = t_t-g_t- \pi \frac{H_{g,t-1}}{p_t}$$

This is our expression for the real terms government deficit. Notice that it has an extra term with respect to the nominal version - the inflation 'loss' term. In this case, despite the minus sign on the last term, it is actually an inflation *gain*. The change in value of the debt is a *positive* for the government. Mathematically, in terms of the equation above, this is true because the multiplicand, $H_{g,t-1}$ in the inflation term is negative in the case of government debt (it is a liability) and so the whole term enumerates as a positive value. 

The private sector's nominal and real budget outcomes are analogously derived. The nominal surplus is trival to derive from the nominal wealth adjustment equation:

$$\Delta H_{h} = H_{h,t} -H_{h,t-1} = Y_t - C_t - T_t $$

That is, the private sector nominal saving is simply the residual left over from income following tax and consumption spending. The real wealth adjustment equation is:

$$h_{h,t} = \frac{H_{h,t}}{p_t}$$

which can be rewritten (by substituting the expression for $H_{h,t}$)

$$h_{h,t} = \frac{H_{h,t-1}}{p_t} +y_t - c_t - t_t$$

and eventually (using similar manipulations as described above),

$$\Delta h_{h} = h_{h,t} - h_{h,t-1} = y_t - c_t - t_t - \pi \frac{H_{h,t-1}}{p_t}$$

Again, the real surplus is similar to the deflated nominal surplus, but with an additional term describing the losses due to inflation.

We're now in a position to plot the sectoral balances.

```python
fig = plt.figure(figsize=(8, 4))

nominal_budget_plot = fig.add_subplot(121, xlim=(1, N), ylim=(-10, 10)) 
nominal_budget_plot.plot(range(N), T-G,   lw=3)                           
nominal_budget_plot.plot(range(N), Y-T-C, lw=3)                         
nominal_budget_plot.grid()                                              
plt.xlabel('time')                                                      
plt.ylabel('nominal budget position')                                   
        
real_budget_plot = fig.add_subplot(122, xlim=(1, N), ylim=(-10, 10))         
real_budget_plot.plot(range(N), t-g-(pi*H_g/p),   lw=3)                        
real_budget_plot.plot(range(N), y-t-c-(pi*H_h/p), lw=3)                      
real_budget_plot.grid()          
real_budget_plot.legend(['government', 'private'])
plt.xlabel('time')                                                      
plt.ylabel('real budget position')                  

plt.tight_layout()
```

Here we have the nominal sectoral balance on the left and the real terms balance on the right. Again, the real terms situation mimics what we have already seen in previous models - the economy converging to a balanced, steady-state in which neither sector is in surplus or deficit. The nominal situation is different. Instead of the nominal government deficit closing through time, it ends up gradually widening, and this is - logically, inevitably - mirrored in the private sector surplus. So we end up with a real terms balanced budget for the government, but a nominal deficit.

The difference between the two is caused by two components: the inflation loss and the price level. Without the inflation loss (actually a gain) term, the real government budget would look to be slightly in deficit. This is equivalent to saying that the private sector would be running a small surplus - that is, net saving. So what this amounts to is that, because of the depreciation in private sector wealth, the private sector needs to continually save a small proportion of real income simply in order to maintain existing levels of real saved wealth. The net saving exactly cancels out the depreciation in the steady-state. Equivalently, the government's real terms excess of spending over taxation is exactly cancelled out by the inflation gains on the government debt.

Let's just finally look at the cumulative balances for the two sectors.

Nominal wealth continues to increase in line with the price level whereas the real purchasing power of this increasing nominal wealth reaches a steady-state. The steady-state value is slightly lower than in previous models, which reflects the slightly lower level og aggregate real income.

Deflating GDP
=============

We can use these formulations to convert nominal quantities into their equivalent "real" quantities. For example, suppose we have the following sequence of annual aggregate income values for a 5 year period:

```python
Y = np.array([100, 104, 110, 112, 115])
Y
```

This suggests that aggregate incomes have grown 15% over the 5 year period.

Now, suppose we had measured the rate of price inflation in as being 2% per annum during the whole time period. If we consider the first year (of income $= 100$) as the reference year, then we can define the year-on-year variation in the price level for the subsequent 4 years using:

$$p_{t} = p_{t-1}(1 + \pi)$$


```python
# initialize array of zeros
p = np.zeros(5)

# set first value to 1 - the reference year
p[0] = 1

# 2% inflation rate
pi = 0.02

# calculation the price level for the subsequent 4 years
for i in [1, 2, 3, 4]:
    p[i] = p[i-1]*(1+pi)

# display
p
```

Okay, so we have a series of price levels for 5 years each representing the effect of a 2% annual inflation rate. Notice that the first year is represented by a price level of 1. This is because this year is our reference year: if we multiply or divide quantities from that year by this price level they remain unchanged. The second year has a price level index of 1.02 which transparently represents just a single year of 2% inflation over the base case price level. Notice, however, how the subsequent years do not simply increase their price index by 0.02. This is because the 2% constant inflation rate (which we have arbitrarily defined) *compounds* through time. 

Now we can use this series of price levels to *deflate* our nominal annual income figures, such that they will be expressed at a common price level using:

$$y = \frac{Y}{p}$$

where $y$ is the *real* value of aggregate income, $Y$ is nominal income, and $p$ is the price level.

```python
y = Y/p
y
```

|--|--|--|--|--|--|
|Y|100|104|110|112|115|
|inflation rate (%)|2|2|2|2|2|
|price level|1|1.02|1.0404|1.0612|1.0824|
|y|100|101.96|105.73|105.54|106.24|


So we've calculated the *real* annual incomes ($y$) for our 5 year period and from these figures it looks as though incomes have grown by ~6% rather than the 15% we'd seen from the nominal values ($Y$). This (if the example were real!) would give us a quite different perspective on the economic growth compared with looking at the nominal figures alone.

The inflation loss term
=======================

There are a few other ways to express this, which we'll explore.

First, we can understand that $H_{h,t-1}$, the previous nominal value of saved wealth, is equal to $h_{h,t-1} \cdot p_{t-1}$, i.e. the previous real value multiplied by the previous price level. So we can rewrite the last equation as:

$$\Delta h_h =  -\pi \frac{p_{t-1}}{p_t} h_{h,t-1}$$

This now expresses the change in real saved wealth entirely in terms of real, rather than nominal, wealth. Since,

$$\frac{p_{t-1}}{p_t} = \frac{1}{1 + \pi}$$

then,

$$\Delta h_h =  - \frac{\pi}{1 + \pi} h_{h,t-1}$$

And making a few more manipulations:

$$\Delta h_h = - \frac{\pi}{1 + \pi} h_{h,t-1}$$

$$\Delta h_h = \frac{1 - 1 - \pi}{1 + \pi} h_{h,t-1}$$

$$\Delta h_h = \frac{1 - (1 + \pi)}{1 + \pi} h_{h,t-1}$$

$$\Delta h_h = \Big ( \frac{1}{1 + \pi} - \frac{1 + \pi}{1 + \pi} \Big ) h_{h,t-1}$$

$$\Delta h_h = \Big ( \frac{1}{1 + \pi} - 1 \Big ) h_{h,t-1}$$

This latter formulation is useful as it resembles something called the [Fisher Equation]() which can be used when interest rates are additionally in play. (In fact, the formulation above is equivalent to the expression for a real interest rate in the special cage that the nominal interest rate is zero).

We can simplify this expression as,

$$\Delta h_h = i \cdot h_{h,t-1}$$

where

$$i = \Big ( \frac{1}{1 + \pi} - 1 \Big )$$

with $i$ representing the coefficient of inflation loss. This term for for describing the change in money stocks due to inflation will be crucial in ensuring the that nominal and real values are dealt with in a stock-flow consistent manner.
