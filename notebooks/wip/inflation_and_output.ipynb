```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

N = 100

# exogeneous variables

Gr = np.zeros(N)   # government spending
Gr[0:50]  = 20
Gr[50:N]  = 20.5

theta = 0.2       # tax rate
alpha_H = 0.2     # propensity to spend out of saved wealth
O_max = 100

# endogeneous variables
G   = np.zeros(N) # nominal g
Y   = np.zeros(N) # nominal income
T   = np.zeros(N) # nominal tax revenue
C   = np.zeros(N) # nominal consumption
H_h = np.zeros(N) # nominal private savings
H_g = np.zeros(N) # nominal government balance


Yr = np.zeros(N) # real income
Tr = np.zeros(N) # real tax 
Cr = np.zeros(N) # real consumption
H_h_r = np.zeros(N) # real wealth
H_g_r = np.zeros(N) # real tax 

S = np.zeros(N) # employment level
I = np.zeros(N) # inventories
P = np.zeros(N) # price level 
```

```python
alpha_Y = np.zeros(N)
alpha_Y[0:10]  = 0.9
alpha_Y[10:50] = 0.85
alpha_Y[50:N]  = 0.85
```

```python
print(alpha_Y[0:15])
```

```python
G[0]  = Gr[0]
Y[0]   =  100
C[0]   =  80
T[0]   =  20
H_h[0] =  40
H_g[0] = -40

Yr[0]  = Y[0]
Tr[0]  = T[0]
Cr[0]  = C[0]
H_h_r[0]  = H_h[0]
H_g_r[0]  = H_g[0]

S[0] = Yr[0]
I[0] = 0 
P[0] = 1 
```

Okay, easy. Let's run the model.

$$Y = \frac {G + \alpha_H H_{t-1}}{1 - \alpha_Y (1 - \theta)}$$

$$y = \frac {Y}{p}$$

$O = $

K = capacity utilisation
O = O_max * K





Period 1
1. 100 output available
2. 100 spent (demand)
3. Everything sold
4. Price level established as 1 (100/100)
5. Firms expect to sell 100 next time period so produce 100 for next time period

Period 2
1. 100 output available
2. 90 spent at current prices
3. 90 sold, 10 left in inventories
4. Price level estalblished as 90/100 (demand/supply)
5. Firms produce 90 (+ 10 left over)

$$Y = \frac {G + \alpha_H H_{t-1}}{1 - \alpha_Y (1 - \theta)}$$
$$y = Y/p$$
$$T = \theta Y$$
$$t = T/p$$
$$G = g*p$$

$$S = y[t-1] + I[t-1]$$
$$O = y$$

$$p = Y/S$$

```python
for t in range(1, N):
    
    G[t] = Gr[t]*P[t-1]
    
    # calculate total income for this time step (equation 1)
    Y[t] = (G[t] + alpha_H*H_h[t-1])/(1 - alpha_Y[t]*(1-theta))
    
    # calculate the tax paid on income for this time step (3)
    T[t] = theta * Y[t]
    
    # calculate the consumption spending for this time step (4)
    C[t] = alpha_Y[t]*(1 - theta)*Y[t] + alpha_H*H_h[t-1]
    Cr[t] = C[t]/P[t-1]
    
    # calculate the new level of private savings for this time step (5)
    H_h[t] = H_h[t-1] + Y[t] - T[t] - C[t]
    H_h_r[t] = H_h[t]/P[t-1]
    
    # calculate the new level of government money balance (6)
    H_g[t] = H_g[t-1] + T[t]- G[t]
    H_g_r[t] = H_g[t]/P[t-1]
   
    
    S[t] = Yr[t-1] + I[t-1]
    P[t] = Y[t]/S[t]
    
    Yr[t] = Y[t]/P[t]
    Tr[t] = T[t]/P[t]
    
    I[t] = (S[t] - Yr[t] if S[t] - Yr[t] > 0 else 0)
```

```python
# initialise plot figure
fig = plt.figure(figsize=(12, 4))

# plot government spending (G) through time
gov_plot = fig.add_subplot(131, xlim=(0, N), ylim=(0, 200))         # set axis limits
gov_plot.plot(range(N), G, lw=3)                       # plot constant G versus time
gov_plot.grid()                                                     # add gridlines
plt.xlabel('time')                                                  # label x axis
plt.ylabel('government spending')                                   # label y axis

# plot consumption spending (C) through time
consumption_plot = fig.add_subplot(132, xlim=(0, N), ylim=(0, 500)) # set axis limits
consumption_plot.plot(range(N), C, lw=3)                            # plot C versus time
consumption_plot.grid()                                             # add gridlines
plt.xlabel('time')                                                  # label x axis
plt.ylabel('consumption')     
# label y axis

# plot aggregate income (Y) through time
income_plot = fig.add_subplot(133, xlim=(0, N), ylim=(0, 500))      # set axis limits
income_plot.plot(range(N), Y, lw=3)                                 # plot Y versus time
income_plot.grid()                                                  # add gridlines
plt.xlabel('time')                                                  # label x axis
plt.ylabel('income')                                                # label y axis

plt.tight_layout() # space subplots neatly
```
