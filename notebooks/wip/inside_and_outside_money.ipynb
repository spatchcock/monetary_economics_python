```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

nt = 100

#G     = 20  # government spending
#theta = 0.2 # tax rate

#alpha_Y = 0.9
#alpha_Y = np.zeros(N) # government spending
#alpha_Y[0:30]  = 0.9
#alpha_H = 0.2

#Y   = np.zeros(N) # income

# Production sector

pr    = 1.0 # productivity (output per person per time period)
W     = 1.0 # wage rate (price per person per time period)
beta  = 0.5 # sales expectation weighting parameter (last sales versus expected last sales)

y     = np.zeros(nt) # output
s     = np.zeros(nt) # realised sales
s_e   = np.zeros(nt) # expected sales
inv   = np.zeros(nt) # realised inventories
inv_e = np.zeros(nt) # expected inventories
inv_T   = np.zeros(nt) # long run inventories target
sigma_T = np.zeros(nt) # long run inventories to sales target ratio
sigma_s = np.zeros(nt) # realised inventories to sales ratio
N     = np.zeros(nt) # employment
WB    = np.zeros(nt) # wage bill
UC    = np.zeros(nt) # unit cost
p     = np.zeros(nt) # price level
NHUC  = np.zeros(nt) # Normal historical unit cost

S     = np.zeros(nt) # nominal realised sales
INV   = np.zeros(nt) # nominal realised inventory level
L_d   = np.zeros(nt) # loans demanded
F_f   = np.zeros(nt) # realised profits of firms

# For random samples from N(\mu, \sigma^2), use:
#  sigma * np.random.randn(...) + mu
s_e_sigma = 0.1
s_e_mu    = 1.0

sigma_0 = 0.1 # base inventories to sales target ratio
sigma_1 = 0.1 # inventories to sales target ratio interest rate modifier  
gamma   = 0.5 # target inventories approach parameter

phi = 0.1 # markup modifier

# Commercial banking sector

r_l  = np.zeros(nt) # interest rate on loans
rr_l = np.zeros(nt) # real interest rate on loans in lieu of inflation
r_m  = np.zeros(nt) # real interest rate on M2 deposits

F_b[t] = np.zeros(nt) # realised profits of banking sector

# Government sector
tau = 0.2 # sales tax

r_b = np.zeros(nt) # real interest rate on bills
g   = np.zeros(nt) # government spending

T  = np.zeros(nt) # realised sales tax revenue

# Household sector
c  = np.zeros(nt) # consumption spending

YD_r  = np.zeros(nt) # regular nominal income
M_2   = np.zeros(nt) # M2 deposits
BB_h  = np.zeros(nt) # bills held

# General
pi = np.zeros(nt) # inflation rate

# Initial conditions

y


```

```python
for t in range(1, nt):
    
    r_l[t] = 0.02    
    
    
    
    # real rate of interest on loans
    rr_l[t] = ((1 + r_l[t])/(1 + pi[t])) - 1 # 10.8
    
    # expected sales
    s_e[t] = beta*s[t-1] + (1 - beta)*s_e[t-1] # 10.5
    # or
    #s_e[t] = s[t-1]*(s_e_sigma * np.random.randn() + s_e_mu) # 10.5A
    
    sigma_T[t] = sigma_0 - sigma_1 * r_l[t] # 10.7
    
    inv_T[t] = sigma_T[t] * s_e[t] # 10.6
    
    inv_e[t] = inv[t-1] + gamma * (inv_T[t] - inv[t-1]) # 10.9
    
    # output, function of expected sales, expected inventory levels and previous inventory levels
    y[t] = s_e[t] + (inv_e[t] - inv[t-1]) # 10.1
    
    # employment
    N[t] = y[t]/pr # 10.2
    
    # wage bill
    WB[t] = N[t]*W # 10.3
    
    # unit cost
    UC[t] = WB[t]/y[t] # 10.4
    
    NHUC[t] = (1 - sigma_T[t]) * UC[t] + sigma_T[t] * UC[t-1] * (1 + r_l[t]) # 10.11
    
    p[t] = (1 + tau) * (1 + phi) * NHUC[t] # 10.10
    
    pi[t]  = (p[t] - p[t-1])/p[t-1]
    
    
    
    
    
    s[t] = c[t] + g[t]
    
    S[t] = s[t] * p[t]
    
    T[t] = S[t] * (tau/(1 + tau))
    
    inv[t] = inv[t-1] + y[t] - s[t]
    
    sigma_s[t] = inv[t-1]/s[t]
    
    INV[t] = inv[t] * UC[t]
    
    L_d[t] = INV[t]
    
    F_f[t] = S[t] - T[t] - WB[t] + (INV[t] - INV[t-1]) - r_l * INV[t-1]
    
    #F_b[t]
    
    YD_r[t] = (F_f[t] + F_b[t]) + r_m[t-1] * M_2[t-1] + r_b[t-1] * BB_h[t-1] + BL_h[t-1]  
    
    
    
    # calculate total income for this time step (equation 1)
    #Y[t] = (G[t] + alpha_H*H_h[t-1])/(1 - alpha_Y[t]*(1-theta))
    
    # calculate the tax paid on income for this time step (3)
    #T[t] = theta * Y[t]
    
    # calculate the consumption spending for this time step (4)
    #C[t] = alpha_Y[t]*(1 - theta)*Y[t] + alpha_H*H_h[t-1]
    
    # calculate the new level of private savings for this time step (5)
    #H_h[t] = H_h[t-1] + Y[t] - T[t] - C[t]
    
    # calculate the new level of government money balance (6)
    #H_g[t] = H_g[t-1] + T[t]- G[t]
   
```
