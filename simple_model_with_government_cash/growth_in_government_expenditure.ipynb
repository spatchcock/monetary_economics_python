```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

Set up numerical scheme

```python
N = 100
```

Exogenous variables
Government spending

```python
growth_rate = 1.0

G  = np.zeros(N)
G[0] = 20

for i in range(1,N):
    G[i] = G[i-1] * (1 + growth_rate/100.0)
```

Exogenous variables
Parameters

```python
theta = 0.2
a_1   = 0.8
a_2   = 0.2
```

Endogenous variables

```python
Y    = np.zeros(N)
YD   = np.zeros(N)
T    = np.zeros(N)
C    = np.zeros(N)
H_h  = np.zeros(N)
H_g  = np.zeros(N)
```

```python

for t in range(1, N):
    # Determine income for this period based on spending in last period
    Y[t] = G[t-1] + C[t-1]
    
    # Now calculate taxation, spending and saving in this period
    T[t]   = theta * Y[t]
    YD[t]  = Y[t]  - T[t]
    C[t]   = (a_1 * YD[t]) + (a_2 * H_h[t-1])
    H_h[t] = H_h[t-1] + YD[t] - C[t]
    H_g[t] = H_g[t-1] + T[t]  - G[t]
   
```

```python

fig = plt.figure()

income_plot = fig.add_subplot(111, xlim=(0, N), ylim=(0, np.max(Y)*1.5))
income_plot.plot(range(N), Y, lw=3)
income_plot.plot(range(N), C, lw=3)
income_plot.plot(range(N), G, lw=3)
income_plot.grid()
```

```python

fig = plt.figure()

gov_plot = fig.add_subplot(111, xlim=(0, N), ylim=(0, np.max(G)*1.5))
gov_plot.plot(range(N), G, lw=3)
gov_plot.plot(range(N), T, lw=3)
gov_plot.grid()
```

```python
fig = plt.figure()

cash_plot = fig.add_subplot(111, xlim=(0, N), ylim=(np.min(H_g)*1.5, np.max(H_h)*1.5))
cash_plot.plot(range(N), H_h, lw=3)
cash_plot.plot(range(N), H_g, lw=3)
cash_plot.grid()
```

```python
fig = plt.figure()

debt_ratio_plot = fig.add_subplot(111, xlim=(0, N), ylim=(0,5))
debt_ratio_plot.plot(range(N-2), (H_g[1:-1]/Y[1:-1])*-1.0, lw=3)
debt_ratio_plot.grid()

fig = plt.figure()

deficit_ratio_plot = fig.add_subplot(111, xlim=(0, N), ylim=(0,1))
deficit_ratio_plot.plot(range(N-2),((T[1:-1]-G[1:-1])/Y[1:-1])*-1, lw=3)
deficit_ratio_plot.grid()
```
